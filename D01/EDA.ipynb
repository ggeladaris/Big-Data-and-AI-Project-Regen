##  How many listings per neighbourhood are there?

```python
df['neighbourhood_cleansed'].value_counts()
```

```python
df['neighbourhood_cleansed'].value_counts().plot(kind = 'bar', figsize = (13, 5))
plt.title('Neighbourhood Distribution', fontsize = 15)
plt.xlabel('Neighbourhood')
plt.ylabel('Number of Listings')
plt.show()
```

## Price based on Room Type

```python
df['price']=df['price'].str.strip('$')
df['price'] = df['price'].apply(lambda x: float(x.split()[0].replace(',', '')))

a = df.groupby('room_type')
# FInd out the mean value of the prices in each room type
prices = a['price'].mean().sort_values(ascending = False)
prices.plot(kind = 'bar', figsize = (10, 5))
plt.title('Price based on Room Type', fontsize = 20)
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

## Price Density Plot

```python
sns.distplot(df.loc[(df.price <400) , 'price'], hist=True, kde=True, 
             bins=int(180/5), color = 'darkred', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
```

## Price Density Plot after Log Transform

```python
sns.distplot(np.log(df.loc[(df.price <400), 'price']), hist=True, kde=True, 
             bins=int(180/5), color = 'darkred', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
```
